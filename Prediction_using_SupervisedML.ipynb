Prediction using Supervised ML
** Predict the percentage of a student based on no. of study hours **
- using a simple linear regression algorithm
- what will be the predicted score if a student studies for 9.25hrs/day?


```python
# importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading data from remote link
url = "https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
student_data = pd.read_csv(url)
print("Data has been imported")

student_data.head(10)

df = pd.DataFrame(student_data)
```

```python
# plotting the data in scatter plot
student_data.plot(x='Hours',y='Scores', style = 'o')
plt.title('Hours with respect to Percentage')
plt.xlabel("Hours studied")
plt.ylabel("Percentage scored")
plt.show()
```

```python
# preparing the data
# dividing them into inputs and outputs

x= df[['Hours']]
y= df["Scores"]
```

```python
# now spliting them into training and test sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.2, random_state=42)
```

```python
# training the algorithm
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
print ('Training is completed.')
```

```python
# plotting the regression line
line = model.coef_*x+model.intercept_

#plotting test data
plt.scatter(x,y)
plt.plot(x,line);
plt.show
```

```python
#make predictions
print(x_test)
y_pred = model.predict(x_test)
print(y_pred)
```

```python
# Evaluate the model
plt.scatter(x_test, y_test, color = 'blue', label = 'Actual percentage')
plt.plot(x_test, y_pred, color='red', linewidth=2, label= 'Predicted Percentage')
plt.xlabel('Hours studied')
plt.ylabel('Percentage scored')
plt.legend()
plt.show()
```

```python
# Predicting the score for given condition : 9.25hrs/day
hrs = [[9.25]]
predicted_score = model.predict(hrs)
print(f'Predicted score for a student studying 9.25 hrs/day: {predicted_score[0]}')
```

```python
# for testing the accuracy of the model, we have 3 measures :
# mean_absolute_error, mean_squared_error, r2_score(R squared score)

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Evaluating the accuracy of the model
mean_ae = mean_absolute_error(y_test, y_pred) #mean absolute error
mean_se = mean_squared_error(y_test, y_pred) #mean squared error
r2 = r2_score(y_test, y_pred) #r2 score

# printing the outputs

print(f'mean absolute error: {mean_ae}')
print(f'mean squared error: {mean_se}')
print(f'r2 score(R_squared score): {r2*100}')
```
